```python
import librosa as lr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
import librosa.display
```

```python
df = pd.read_csv('drive/MyDrive/Colab_Notebooks/project/UrbanSound8K/metadata/UrbanSound8K.csv')
```

```python
def create_spectrogram(y):
    spec = lr.feature.melspectrogram(y=y)
    spec_conv = lr.amplitude_to_db(spec, ref=np.max)
    return spec_conv
```

```python
for i in range(0,8731):
    y, sr = lr.load('drive/MyDrive/Colab_Notebooks/project/UrbanSound8K/audio/fold{}'.format(df.fold[i])+'/{}.wav'.format(df.slice_file_name[i].strip(".wav'")))
    spec = create_spectrogram(y)
    fig = plt.Figure()
    canvas = FigureCanvas(fig)
    ax = fig.add_subplot(111)
    p = lr.display.specshow(spec, ax=ax)
    fig.savefig('drive/MyDrive/Colab_Notebooks/project/spectrogram/{}'.format(df.classID[i])+'/{}.png'.format(df.slice_file_name[i].strip(".wav'")))
```
